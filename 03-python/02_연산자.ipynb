## 비교 연산자
|연산자|내용|
|---|---|
|`>`, `<`|초과, 미만|
|`>=`, `<=`|이상, 이하|
|`==`|값이 같음|
|`!=`|다르다|
|`in`|포함여부|
|`is`|동일하다|

```python
# 크고 작음

5 > 3  # T
4 < 2  # F
4 >= 4  # T
3 <= 1  # F
```

```python
5 == 3  # F
4 != 2  # T
'apple' == 'banana'
```

```python
# x in y -> y안에 x가 있는지

'i' in 'apple'  # False
3 in [1, 2, 3, 4]  # True
```

```python
# bmi 계산기
# bmi = 몸무게(kg) / 키(m) 제곱 2
# weight 에 몸무게, height 에 키를 저장하고 bmi 를 계산하기

weight = 80
height = 1.8

bmi = weight/(height**2)

print(bmi)
```

```python
# 원 넓이와 둘레 계산 (pi 는 3.14 로 계산)
# 원 넓이 = pi x r^2
# 원 둘레 = 2 x pi x r
r = 3
pi = 3.14
square = pi * r**2
circumference = 2 * pi * r

print(square, circumference)
```

```python
import math
```

```python
# 초를 시간, 분, 초 로 바꾸기
# 계산을 통해 h, m, s 에 시간, 분, 초로 전환

seconds = 3675  # 1시간 1분 15초

minute = seconds / 60
print(minute)

hour = int(math.floor(minute / 60))
print(hour)

minute = int(math.floor(minute % 60))
print(minute)

second = int(math.floor(seconds % 60))
print(second)

print(f'{hour}시간 {minute}분 {second}초')
```

```python
# 세자리 수 분리하기
# 숫자를 백, 십, 일의 자리로 분리하여 변수에 저장하기

n = 472

a = int(math.floor(n/100))
b = int(math.floor(((n%100)/10)))
c = int(math.floor(((n%100)%10)))
print(a,b,c)
```
