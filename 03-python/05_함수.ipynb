````python
# name: plus_one / in: 숫자 / out: 숫자 + 1

# 함수 정의 (define)
def plus_one(num):
    return num + 1

# 함수 호출 (call)
plus_one(99)
````

````python
# n번 인사하는 함수
# name: hello_ntimes

def hello_ntimes(n):
    # for-매개변수에 들어간 값을 사용하지 않을 때 => _
    for _ in range(n):
        print('반가워')

hello_ntimes(5) # 5번 hello를 출력
````

````python
# 3제곱해주는 함수
# name: cube, in: 숫자 1개, out: 숫자의 세제곱

def cube(n):
    return n ** 3
        
cube(3) # 27
````

````python
# 원기둥의 부피를 구하는 함수
# name: cylinder, in: 반지름/높이, out: 부피

pi = 3.14159
def cylinder(r,h):
    return r**2 * pi * h

cylinder(2,5)
````

````python
# 리스트 내부의 숫자들의 평균을 구해주는 함수
# name: get_avg, in: 숫자들로 만들어진 리스트, out: 평균

def get_avg(lists):
    total_sum = 0
    average = 0
    for list in lists:
        total_sum += list
    average = total_sum/len(lists)
    return average

get_avg([1,2,3,4,5]) #3
````

````python
# 비밀번호 체크해주는 함수 (숫자 포함, 8글자 이상)
# name: check_pw, in: 비밀번호 문자열, out: True/False

def check_pw(password):
    character_number_check = False
    if len(password) >= 8:
        for character in password:
            if character in '0123456789':
                character_number_check = True
    
    if len(password) >= 8 and character_number_check == True:
        print(True)
    else:
        print(False)

check_pw('abcd1234') # True
check_pw('abcdefg') # False
````

## 함수
1. 내장함수 -> 즉시 호출 (print, bool, type, len)
2. 외장함수 -> 다운로드된 거 호출 (math)
3. 라이브러리 -> 직접 외부에서 다운로드 받아야함 (pandas)

````python
# BMI 기준으로, 비만여부 확인해주는 함수 (체중/키(m)^2)

def check_bmi(weight, height, gender):
    BMI = weight/((height/100)**2)
    if gender == 'M':
        if BMI >= 27:
            return '비만'
        else:
            return '정상'
    elif gender == 'F':
        if BMI >= 25:
            return '비만'
        else:
            return '정상'


print(check_bmi(100, 175, 'M')) # 남성
print(check_bmi(50, 160, 'F')) # 여성
````
