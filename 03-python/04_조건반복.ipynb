## 조건
- if - else

### if 구문

```python
# 조건 - IF

n = -10

if n > 0:
    # 블럭 안은 조건이 True 일때만 실행
    print('양수 입니다')
    print('11111')

# 블럭 밖은 조건문과 상관 없음
print('22222')  
```

```python
# 조건 - IF-ELSE

n = 10

if n > 0:
    print('0보다 크다.')
else:
    print('0보다 작거나 같다.')
```

```python
# 조건 - IF-ELIF-ELSE

n = -10

if n > 0:
    print('양수')
elif n == 0:
    print('0이다')
elif n < 0:
    print('음수')
```

```python
score = 765

# 불합격: 0-399, LV.1: 400-599, LV.2: 600-749, LV.3: 750-899, LV.4 900-1000

if score <= 1000 and score >= 900:
    print('LV.4')
elif score < 900 and score >= 750:
    print('LV.3')
elif score < 750 and score >= 600:
    print('LV.2')
elif score < 600 and score >= 400:
    print('LV.1')
else:
    print('불합격')
```

```python
# 우수 학생 판별
status = {
    "score": 80,
    "absent": 7
}

# 1. 점수가 70을 넘는 학생이면 우수학생
# 2. 우수학생 중에 결석일이 5일 미만이면 최우수학생
# 3. (우수 아닌) 결석일이 5일 미만이면 성실학생
# 4. 나머지는 일반학생

if status['score'] >= 70:
    if status['absent'] < 5:
        print('최우수학생')
    else:
        print('우수학생')
elif status["absent"] < 5:
    print('성실학생')
else:
    print('일반학생')
```

## 반복
- while
- for

### while 구문

```python
n = 0
while n < 5:
    print('출력', n)
    n += 1

print('끝나고', n)
```

```python
numbers = [1, 2, 3, 4, 5]

total = 0
index = 0

while index < len(numbers):
    total += numbers[index]
    index += 1


print(total)
```

### for .. in 구문

```python
numbers = [1, 2, 3, 4, 5]
total = 0

for num in numbers:
    total += num

print(total)
```

```python
for n in range(100):
    print(n)
```

```python
# 리스트 index 수정 가능, 문자열은 index 수정 안됨
s = 'python'
s[0:3] # 조회 가능
# s[0] = 'P' # 수정 불가능

for c in s:
    print(f"!{c}!")
```
