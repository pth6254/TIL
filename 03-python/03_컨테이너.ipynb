## 리스트 (List)

```python
 # 리스트 이름 -> 복수형으로 이름 짓기

# index는 0부터 (리스트 길이 -1) 까지
numbers[4]
# index -1부터 -리스트길이
numbers[-2]
```

```python
numbers = [10, 20, 30, 40, 50]

# 슬라이싱 [x:y] -> x부터 y전까지
numbers[0:3] # 0~2
numbers[1:1] # x

numbers[0:5] # 0~4 전체
numbers[:3] # 2까지
numbers[2:] # 2부터 
numbers[:] # 처음부터 끝까지

# 값은 같다
numbers == numbers[:]
# 같은 존재냐?
numbers is numbers[:]
```

```python
# 리스트 수정
numbers = [10, 20, 30, 40, 50]

# 특정 index의 데이터 바꾸기
numbers[2] = 300
numbers[-1] = 500
print(numbers)

# 리스트에 데이터 추가하기
numbers.append(60)
print(numbers)

# 리스트의 데이터 삭제하기
numbers.pop(-1)
print(numbers)
```

## 범위 (Range)

```python
# range는 처음부터 마지막 -1까지의 정수만을 의미
r1 = range(1, 10)
# 눈으로 보려면, 리스트로 바꿔줘야함
print(list(r1))

2.4 in r1

# 시작숫자 x -> 0부터 시작
r2 = range(5)
print(list(r2))

# step 개념
r3 = range(0, 10, 2)
print(list(r3))

# 음수 step(뒤로가기) -> 시작, 끝, 방향+스텝
r4 = range(5, 0, -1)
print(list(r4))
```

## Dict (딕셔너리)
| Key - Value
- JSON의 뜻은 JavaScript Object Notation의 줄임말
    - JS언어의 Object 자료형 식 표기법 (" " 사용 필수) 
- Python에는 Dict 자료형

```python
# 생성

me = {
    "name": "neo",
    "age": 20,
    "is_hungry": True,
    "height": 123.4,
    "skills": ['n8n', 'sql', 'python']
}
```

```python
# 접근 -> 대괄호 안에 key를 쓴다
me["name"], me["age"]

# 'python'에 접근
me["skills"][2]
```

```python
# 변경
me["name"] = "김새싹"
print(me)

me["is_hungry"] = False
print(me)
```

```python
# me["skills"]는 리스트! 리스트에 아이템 추가는 append 함수!
me["skills"].append('git')
print(me)
```

```python
# K - V 추가
me["money"] = 10000
print(me)

# K - V 삭제
me.pop("height")
print(me)
```

```python
print(type(me))
print(type(me["name"]))
print(type(me["is_hungry"]))
print(type(me["skills"]))
```

## Set (집합)
- 중복되는 요소가 없다.
- 순서(index)가 없다.

```python
s1 = {1, 1, 1, 2, 3, 3, 3}
s2 = {3, 4, 5, 5}
print(s1)
print(s2)

# 합집합
print(s1 | s2)
# 교집합
print(s1 & s2)
# 차집합
print(s1 - s2)

# 집합 -> 리스트 변환
list(s1)

# 중복이 많은 리스트에서 중복을 제거하는 용도로 많이 쓰임
ll = [1, 1, 1, 1, 2, 2, 2, 3, 4, 5, 5, 5, 5, 6]
ss = set(ll)
print(ss)
ll = list(ss) # 리스트로 바뀌었을 때의 순서는 의미 없음
ll.sort() # 오름차순 정렬을 할 수 있음
print(ll)
```

## Tuple (튜플)
- 리스트 친척
- 순서(index)가 있음
- 리스트와 차이: 수정 불가능
- 직접 쓰기보다 python 내부에서 많이 쓰임

```python
t = (10, 20, 30, 40, 50)
# 접근가능
print(t[0:3], t[-1], t[:])
```

```python
# 수정불가능 (실행시키면 튜플이라서 실행 못한다는 에러뜸)
t[0] = 100
t.append()
t.pop()
```
