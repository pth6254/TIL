```python
# Person
# 특성: 키, 몸무게, 나이, 이름
# 행동: 인사, 공부

# python에서 클래스 이름은 PascalCase로 짓는다 (예시> 일반 변수: my_good_friend / 클래스: MyGoodFriend)
class Person: # 개념
    # 1. 특성 (attribute) -> 변수
    height = 165
    weight = 55
    age = 20
    iq = 100
    name = '인간'
    # 2. 행동 (method) -> 함수
    def greeting(self):
        return f'안녕하세요, {self.name}입니다. {self.age}살 입니다.'
    def study(self):
        self.iq += 5
```

```python
# Person 개념적 대상을 -> 실제 생성
p1 = Person()
p2 = Person()
```

```python
# 원 (Circle)
class Circle:
    # 생성자함수 -> 처음 인스턴스가 만들어질 때 자동 호출되는 함수
    def __init__(self, r):
        self.r = r
        print(f'반지름이 {self.r}인 원이 생성되었습니다.')
    
    def get_area(self):
        return 3.14 * self.r**2
    
    def __add__(self, other):
        return Circle(self.r + other.r)

c1 = Circle(3) # 반지름이 3인 원
c2 = Circle(5)

print(c1.get_area()) # 반지름이 3인 원의 넓이
print(c2.get_area())

c3 = c1 + c2 # r=3 + r=5 -> r=8 원을 새롭게 만들겠다
c3.get_area()
```
